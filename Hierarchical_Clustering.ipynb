**Hierarchical Clustering - Intuition**  
* There are 2 types: Agglomerative & Divisive = Bottom Up & Top Down  
* We will focus on Agglomerative  
* STEPS:  
1. Make each data point a single point cluster = N clusters  
2. Take 2 closest data points and make them one cluster = (N-1) clusters  
3. Combine 2 closest = (N-2) clusters  
4. Repeat until only 1 cluster  
* Key part of algorithm is how you define definition between 2 clusters  
* Examples:  distance between centroids ; distance between closest points ; etc.  
* Resulting progression is stored in a dendogram  

**How do Dendrograms work?**  
* All data points are listed on the x-axis  
* Distance is the y-axis  
* Horizontal line joins points that are combined at each step above  
* Vertical distances of 2 endpoints below horizontal represents Euclidean distance between points  
* Further the vertical distance means more dissimilar the join  
* Join individual points on the bottom - Join clusters at the top center of horizontal  
* Joining point (bottom) to a cluster (horizontal) - 2 legs unequal - distance is longer line (from bottom)  

**Dendrogram Intuition**  
* Set similarity thresholds via vertical distances  draw a vertical line at that level  
* However many vertical lines are crossed represents the number of clusters  
* For each vertical that is crossed - corresponding horizontals represent clusters  
* Extend each horizontal ; choose longest vertical that is not intersected by horizontal that it isn't anchored on

```r
# Import Mall Dataset  
data.mall <- read.csv('Mall_Customers.csv')

X <- data.mall[,4:5]

# Use dendrogram to find optimal number of clusters
dendrogram <- hclust(dist(X, method = 'euclidean'),
                    method = 'ward.D')

plot(dendrogram,
    main = paste('Dendrogram'),
    xlab = 'Customers',
    ylab = 'Euclidean Distances')
```

* Analyzing the above - results in 5 clusters

```r
hc <- hclust(dist(X, method = 'euclidean'),
                    method = 'ward.D')

y_hc <- cutree(hc, 5)
```

```r
# Visualize the clusters
library(cluster)

clusplot(X,
        y_hc,
        lines = 0,
        shade = TRUE,
        color = TRUE,
        labels = 2,
        plotchar = FALSE,
        span = TRUE,
        main = paste('Clusters of Clients'),
        xlab = 'Annual Income',
        ylab = 'Spending Score')
```

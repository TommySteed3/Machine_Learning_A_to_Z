**K-MEANS**  
* Clustering algorithm.    
* STEPS:  
1. Choose k - the number of clusters.  
2. Select a random k points (centroids) - not necessarily from your dataset.  
3. Assign each datapoint to the closest centroid.  This forms k clusters.  
4. Compute and place the new centroid of each cluster.  
5. Reassign each dataset to the new closest centroid.  Repeat step 4 until no more movement of centroids.  

**K Means Random Initialization Trap**  
* "Bad" initial centroids can affect results.  
* K-Means++ selects a more efficient initial centroid.  
* R and Python has this built in to their algorithms.  

**Selecting the number of clusters**  
* WCSS (Within Cluster Sum of Squares): Sum distance to center squared for all data points and all clusters.  
* Decreases as you increase clusters.  
* Graph WCSS x Clusters - and look for elbow.  Becomes basically linear where you can stop.  
* Arbitrary - some datasets may not be as obvious as others.  

```r
# Import mall dataset
data.mall <- read.csv('Mall_Customers.csv')
```

```r
X <- data.mall[,4:5]
```

```r
# Elbow method to select optimal number of clusters.  Use a loop.
set.seed(6) 
wcss <- vector()
for(i in 1:10) wcss[i] <- sum(kmeans(X,i)$withinss)
plot(1:10,wcss,type = 'b', main = paste('Clusters of clients'), xlab = 'Number of Clusters', ylab = 'WCSS')
```

```r
# Apply k-means to mall dataset
set.seed(29)
kmeans <- kmeans(X, 5, iter.max = 300, nstart = 10)
```

```r
# Visualize the clusters
library(cluster)

clusplot(X,
        kmeans$cluster,
        lines = 0,
        shade = TRUE,
        color = TRUE,
        labels = 2,
        plotchar = FALSE,
        span = TRUE,
        main = paste('Clusters of Clients'),
        xlab = 'Annual Income',
        ylab = 'Spending Score')
```
